```python
# Import txt file

with open('20C_events_article_Wiki.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', '')
```

```python
import sys
print(sys.executable)
```

```python
from nltk.tokenize import word_tokenize
print(word_tokenize("This is a test."))
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# Frequency distribution
from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words)
```

```python
# Frequency Distribution Plot of 10 most common words
plt.figure(figsize=(8, 3))
dist_words.plot(10,
                cumulative = False)
plt.show()
```

Removing stopword and punctuation marks

```python
# Defining stopwords

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
# Removing stopwords in words

filtered_words = [] # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
# Create a new FreqDist for filtered_words

dist_words_filter = FreqDist(filtered_words)
print(dist_words_filter)
```

```python
# Frequency Distribution Plot of filtered words

plt.figure(figsize=(8, 3))
dist_words_filter.plot(10, cumulative = False)
plt.show()
```

```python
# Substitute all punctuations marks with a space 

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(filtered_words))
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print(tokenized_word_2)
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
# Frequency Distribution Plot after removing punctuation marks and stopwords

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(30, cumulative = False)
plt.show()
```

We can now get a feel for the main topics covered by the article (war between certain nations). In the earlier plots (before removing stopwords and punctuations), you couldn't really learn anything about what the article was about from looking at the most frequent words

## POS Tagging

```python
# first removing more stopwords

new_stopwords = ["th", "Then", 'In', 'S', 's', 'The', 'ISSN', 'ISBN']

filtered = []
for word in tokenized_word_2:
     if word not in new_stopwords:
        filtered.append(word)
```

```python
# make one big string out of filtered list and pass to TextBlob() function
%%time
text = TextBlob(str(filtered))
```

```python
tags_list = text.tags
```

```python
tags_list
```

```python
# store tags in df and view top20
df_text = pd.DataFrame(tags_list)
df_text.columns = [ 'Words', "Word type"]
df_t=df_text.groupby('Word type').count().reset_index()
top10=df_t.nlargest(10, 'Words')
```

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 20):
      sns.barplot(x = "Words", y = "Word type",
     saturation = 0.9, data = top10).set_title("wikipedia 20th Century Events - top 10 word types used")
```

Nouns

```python
df_n = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNP")]
df_n.columns = ["Word", "Occurences"]
x = df_n.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15_n = y.nlargest(15, 'Occurences')
```

```python
plt.figure(figsize=(10, 5))

ax = sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top15_n)
ax.set_title("Wikipedia 20C Events - most frequently used nouns")
plt.xticks(rotation=45)
plt.show()
```

Verbs

```python
df_v = df_text[(df_text['Word type'] == "VB")  | (df_text['Word type'] == "VBD")]
df_v.columns = ["Word", "Occurences"]
x = df_v.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15_v = y.nlargest(15, 'Occurences')
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(x="Word", y="Occurences", 
                 saturation=0.9, 
                 data=top15_v)
ax.set_title("Wikipedia 20C Events - most frequently used verbs")
plt.xticks(rotation=45)
plt.show()
```

Adjectives

```python
df_a = df_text[df_text['Word type'] == "JJ"]
df_a.columns = ["Word", "Occurences"]
x = df_a.groupby('Word').count().reset_index()
y = x.sort_values(by=['Occurences'], ascending=False)
top15_a = y.nlargest(15, 'Occurences')
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(x="Word", y="Occurences", 
                 saturation=0.9, 
                 data=top15_a)
ax.set_title("Wikipedia 20C Events - most frequently used adjectives")
plt.xticks(rotation=45)
plt.show()
```

from the three plots, it's clear to see that the article focuses on wars and armed conflict, especially between the wealthy industrialised nations

## Counting mentions of different countries

```python
listToStr = ' '.join([str(elem) for elem in filtered])
 
print(listToStr)
```

```python
# Create a count for the countries

all_counts = Counter(re.sub(r'\W+', ' ', listToStr).split()) 
```

```python
countries = pd.read_csv("countries.csv")
countries.head()
```

```python
countries_list = countries['Country'].to_list()
```

```python
dict_of_counts = {d : all_counts[d] for d in countries_list}
```

```python
#Search for the names from the list in the dictionary

dct = {v:[k] for v,k in dict_of_counts.items()}
df = pd.DataFrame(dct)
```

```python
df
```

```python
df = df.transpose().reset_index().copy()
```

```python
df.rename(columns = {"index":"Country", 0:"Times mentioned"}, inplace = True)
df.head()
```

```python
df['Times mentioned'].value_counts()
```

```python
# filtering for countries that were mentioned at least once
df = df[df['Times mentioned'] >= 1].copy()
```

```python
plt.figure(figsize=(15,10))
ax = sns.barplot(data=df.sort_values('Times mentioned', ascending=False), 
                 x = 'Times mentioned', 
                 y = 'Country')
plt.show()
```
