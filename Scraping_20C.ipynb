```python
# Import libraries

import pandas as pd
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import matplotlib.pyplot as plt 
import os
import logging
from bs4 import BeautifulSoup
import requests 
```

```python
# Install Driver for webscraping
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
```

```python
# get url
page_url = requests.get("https://en.wikipedia.org/wiki/Key_events_of_the_20th_century")
```

```python
# Create soup and get title

soup = BeautifulSoup(page_url.text, 'html.parser')
print(soup.title)  
```

```python
# Create text object with contents
text = soup.get_text()

text = text.encode ('utf-8')

with open('20C_events_article_Wiki.txt', 'wb') as f:
       f.write(text)
```

```python
# Scraping list of country names 

page_url_countries = 'https://simple.wikipedia.org/wiki/List_of_countries'
driver.get(page_url_countries)
```

```python
# Target all <a> tags inside <p> tags, inside the main content area
country_elements = driver.find_elements(By.CSS_SELECTOR, "div.mw-parser-output > p > a")

# Extract country names
countries = [el.text for el in country_elements]
```

```python
for country in countries:
    print(country)
```

```python
df = pd.DataFrame(countries, columns=["Country"])
df.head(10)
```

```python
# Dropping first row which captured the section heading eroneously 
df.drop(df[df['Country'] == 'sovereign states'].index, 
        inplace=True)
df.head(10)
```

```python
# saving as csv
df.to_csv("countries.csv", index=False)
```
